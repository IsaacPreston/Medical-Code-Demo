```python
grouped_data = data.groupby(['age', 'sex'])['weight'].mean()
```

```python
grouped_data.plot(kind='bar')
plt.xlabel('Age')
plt.ylabel('Mean Weight')
plt.show()
```

```python
clean_data = data[data['weight'] <= 200]
```

```python
data.plot()
```

```python
smokers_data = data.loc[data['smoke'] == 1, 'weight']
plt.hist(smokers_data)
plt.xlabel('Weight')
plt.ylabel('Frequency')
plt.title('Distribution of Weight for Smokers')
plt.show()
```

```python
counts = data.groupby(['sex', 'cholesterol']).size().unstack()
counts.plot(kind='bar', stacked=True)

plt.xlabel('Sex')
plt.ylabel('Count')
plt.title('Count of Cholesterol by Sex')
plt.show()
```
